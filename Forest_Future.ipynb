```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
df = pd.read_csv("/content/cleaned_deforestation_india.csv")
X = df[['Year (since 2000)']]
y = df['Forest Cover Change (sq km)']
model = LinearRegression(fit_intercept=False)
model.fit(X, y)
y_pred = model.predict(X)
plt.scatter(df['Year (since 2000)'], df['Forest Cover Change (sq km)'], color='blue', label='Data Points')
plt.plot(df['Year (since 2000)'], y_pred, color='red', linewidth=2, label='Regression Line (Through Origin)')
plt.xlabel("Year (since 2000)")
plt.ylabel("Forest Cover Change (sq km)")
plt.title("Deforestation Trend in India (Linear Model Passing Through Origin)")
plt.legend()
plt.grid(True)
plt.show()
```
